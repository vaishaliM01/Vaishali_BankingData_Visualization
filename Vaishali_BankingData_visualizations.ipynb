# Bar Plots for Categorical column "job" with target "termdeposit" as hue

```python
# Unique values of job
df['job'].unique()
```

```python
fig = plt.figure(figsize=(10, 6)) 
pd.crosstab(df['job'],df['termdeposit']).plot(kind = 'bar')
```

People with jobs like "admin","blue-collar" and "technician" tend to open more term deposits.

# Bar Plots for Categorical column "marital" with target "termdeposit" as hue

```python
fig = plt.figure(figsize=(10, 6)) 
pd.crosstab(df['marital'],df['termdeposit']).plot(kind = 'bar')
```

Married people have more term deposits.

# Based on education
